The main aim of our EDA is to familiarize ourselves with data as we analyse White Wine data. We'll try to conclude what factors affect the quality of wine.

It has 12 variables.Shape of the dataset is (4898, 12).

## Attribute information:

   For more information, read [Cortez et al., 2009].

   Input variables (based on physicochemical tests):
   1. fixed acidity (tartaric acid - g / dm^3)
   2.  volatile acidity (acetic acid - g / dm^3)
   3.  citric acid (g / dm^3)
   4.  residual sugar (g / dm^3)
   5.  chlorides (sodium chloride - g / dm^3
   6.  free sulfur dioxide (mg / dm^3)
   7.  total sulfur dioxide (mg / dm^3)
   8.  density (g / cm^3)
   9.  pH
   10.  sulphates (potassium sulphate - g / dm3)
   11.  alcohol (% by volume)
   
   Output variable (based on sensory data): 
   12.  quality (score between 0 and 10)

### Bivariate Analysis

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap="YlGnBu",annot=True)
```

Above heatmap tells us that density and residual sugar are strongly related, total sulfur dioxide and free sulfur dioxide are strongly related, and alcohol and density are negatively correlated.
Quality has a relation with alcohol.

**Quality vs Alcohol**

The graph between **quality** and **alcohol** content level is ploted below:

```python
plt.figure(figsize=(10,6))
sns.regplot(x='quality', y='alcohol', data=df, x_jitter=0.3, scatter_kws={'alpha':0.2})
sns.despine()
```

Inferences are:

1. Alcohol and quality are correlated with each other with a correlation coefficient of 0.44
2. Higher the alcohol content (% by volume) in the wine, better is the quality of wine observed. However, *correlation doesn't imply causation*.
3. Averege quality wine samples have alcohol content in mediocre quantities. As wines rated as quality rating 5 have their alcohol content ranging between 8.5 - 11 and wine samples given quality rating 6 have their alcohol content ranging between 8.5 - 13.
4. Wine samples rated 7 have some samples having alcohol content around 9 and 10, bust most of the samples have alcohol content greater than 10.
5. Higher qauality wine samples (rated 8 and 9) have greater alcohol content, where qaulity rating 8 samples are around 11-13 for most of the samples and quality rating 9 samples are few in number but have alcohol content around 12 and 13.

**Quality vs Density**

The graph between **quality** and **density** is plotted below

```python
sns.regplot(x='quality', y='density', data=df, x_jitter=0.3, scatter_kws={'alpha':0.2})
```

We can infer following points from the above plot:

1. **quality** and **density** are negatively correlated with each other with a correlation coefficient of -0.31
2. Top rated wine samples (quality rating of 9) are observed to have lower densities of aroun 0.99 g / cm^3.
3. Average quality wine samples have densities of 0.99, 1.00 and some have greater than 1.
4. Higher the wine quality, lower is the density observed.

**Alcohol vs Density**

The graph between **alcohol** and **density** is plotted below

```python
plt.figure(figsize=(10,6))
sns.regplot(x='alcohol', y='density', data=df, x_jitter=0.2, scatter_kws={'alpha':0.15})
sns.despine()
```

Inferences which can be drawn from above plot are:

1. **density** and **alcohol** are correlated with a correlation coefficient of -0.78
2. As the alcohol content increases the density of wine is observed to decrease.
3. Wines with Higher alcohol concentration (13 and 14) are observed to have lower densities aroun 0.99 g / cm^3
4. Wines containing lower alcohol content are observed to have greater densities.
5. There are some outliers too.

**Alcohol vs Residual sugar**

The graph between **alcohol** and **residual sugar** is ploted below:

```python
plt.figure(figsize=(10,6))
sns.regplot(x='alcohol', y='residual sugar', data=df, x_jitter=0.2, scatter_kws={'alpha':0.15})
sns.despine()
```

Inference drawn from the above plot are:

1. **residual sugar** and **alcohol** are correlated with a correlation coefficient of -0.45
2. Higher the alcohol content in the wine, lower is the residual sugar observed.
3. Wine samples having higher alcohol level have negligible residual sugar level in them.
4. Wines having moderate alcohol levels have around 0 - 15 g / dm^3 of residual sugar in them.
5. Wines having lower alcohol levels have higher amount of residual sugar in them.
6. There are outliers present, as it can be seen from the plot.

### Multivariate Analysis

```python
grid = sns.FacetGrid(df, col='quality',col_wrap = 2)
grid.map(plt.scatter,'alcohol','density',alpha = 0.2)
```

Much can't be inferred from the above plot in terms of denity.

However, alcohol content is in a positive correlation with quality of wine.

```python
grid = sns.FacetGrid(df, col='quality',col_wrap = 2)
grid.map(plt.scatter,'alcohol','total sulfur dioxide',alpha = 0.2)
```

It can possibly be inferred from the above plot that the wine samples having lower **total sulfur dioxide** content and higher **alcohol** content are of better **quality**

```python
grid = sns.FacetGrid(df, col='quality',col_wrap = 2)
grid.map(plt.scatter,'alcohol','free sulfur dioxide',alpha = 0.2)
```

Again, lower concentrations of **free sulfur dioxide** and higher **alcohol** levels are observed to contribute towards better quality of wine.

From the above two plots, we can infer that wine samples having lower sulfur dioxide(total and free) and having pH in a range of 3 - 3.5 are rated higher in quality.
